```python
# 🚀 STEP 1: Install required packages
!pip install diffusers transformers accelerate safetensors

# 🚀 STEP 2: Load the Stable Diffusion model (no login needed)
from diffusers import StableDiffusionPipeline
import torch

model_id = "CompVis/stable-diffusion-v1-4"

# Load the pipeline (fp16 revision for GPU use)
pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    revision="fp16",
    torch_dtype=torch.float16
)
pipe = pipe.to("cuda")  # use GPU for fast generation

# 🚀 STEP 3: Define the text-to-image generation function
def generate_image(prompt):
    image = pipe(prompt).images[0]
    return image

# 🚀 STEP 4: Get user input and generate the image
prompt = input("Enter a prompt for image generation: ")
image = generate_image(prompt)
image.show()
```
